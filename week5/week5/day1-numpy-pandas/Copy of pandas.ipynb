##Data Combination
 Data combintion is the process of joining,merging or concatenating multiple pandas data structures into a single data structure.
  The most common ways of combine dataframes are concatenation,joining and merge
  

## Concatenation

using the concat function to combine 2 or more dataframes with the same column either vertically or horizontally.

```python
df_1=pd.DataFrame({'a':[4,2,6],'b':['m','n','o']})
df_2=pd.DataFrame({'a':[45,8,9],'b':['x','y','z']})
```

```python
df_1
```

```python
df_2
```

```python
#vertical concatenation(row.wise)
vertical_com=pd.concat([df_1,df_2], axis=0, keys=['df_1','df_2'])
vertical_com
```

```python
# Horizontal concatenation is a rowise concatenation.
horizontal_com =pd.concat([df_1,df_2],axis=1,keys=['df_1','df_2'])
horizontal_com
```

```python
#loc is location
vertical_com.loc['df_1',0]
```

```python
vertical_com.loc["df_1",0][0]
```

```python
vertical_com.loc["df_1",0].iloc[0]
```

```python
vertical_com.loc["df_1",0]["a":"b"]
```

```python
#vertical concatenation (row-wise)
vertical_com = pd.concat([df_1,df_2], axis=0,ignore_index=True)
vertical_com
```

The merge function is used to conbine two or more dataframes based on a common column(s)

```python



df_5=pd.DataFrame({'Names':['Ama','Berry','Dela','Tola'],'Score1':[1,2,3,4]})
df_6=pd.DataFrame({'Names':['Berry','Moses','Debby','Dela'],'Score2':[5,6,7,8]})
```

```python
df_5
```

```python
df_6
```

```python
pd.merge(df_5,df_6, on='Names', how='inner')
```

```python
pd.merge(df_5,df_6, on='Names',how='outer')
```

```python
pd.merge(df_5,df_6,on='Names',how='left')
```

```python
pd.merge(df_5,df_6,on='Names',how='right')
```

```python
# The two dataframeshaving different column names
df_7=pd.DataFrame({'Name1':['Ama','Berry','Celestine','Dela'],'Score':[1,2,3,4]})
df_8=pd.DataFrame({'Name2':['Berry','Dela','Emma','Frank'], 'Score':[5,6,7,8]})
merge_df=pd.merge(df_7,df_8,left_on='Name1',right_on='Name2', how='inner',suffixes=('_df7','_df8'))
merge_df
```

The join function id used to combine two dataframes on their indexes

```python


df_9=pd.DataFrame({'Value1':[1,2,3,4]}, index=['A','B','C','D'])
df_10=pd.DataFrame({'Value2':[5,6,7,8]},index=['B','D','E','F'])
```

```python
df_9
```

```python
df_10
```

```python
df_9.join(df_10, how='inner')
```

```python
joined=df_9.join(df_10, how='left',lsuffix='_left',rsuffix='_right')
joined
```

Grouping and aggregation

You can group data based on one or more columns and perform aggregation functions.

```python
# Creating a Dataframe from a dictionary
data={'Num':[10,20,30],
      'Age':[10,25,35],
      'val':[50,60,70]}
df=pd.DataFrame(data)
df
```

```python
group_df=df.groupby('Age')
group_df
```

```python
group_df=df.groupby('Age').mean()
group_df
```

```python
group_df=df.groupby('Age').max()
group_df
```

Reading files with Pandas:

Import pandas:
Import the pandas library in your python script or jupyter Notebook.
Import pandas as pd
Readin CSV files:
Use the read_csv() function to read from the CSV(Comma-seperated Values) file.
Syntax
df=pd.read_csv('filename.csv')
options:
sep: Specifies the delimeter(default is,).
header:Specifies the row number to use as column names.
index_col:Specifies the column to use as the row index.
usecols: Specifies a subset of column to read.


```python
import pandas as pd
# Reading CSV FILE WITH A DIFFERENT DELIMETER
df1 = pd.read_csv('sample_data.csv', sep=',')
print(df1.head())  # Displaying the first few rows of the DataFrame
```

```python
import pandas as pd

# File path of the Excel file
excel_file_path = "/content/sample_data_excel.xlsx"

# Read the Excel file using pandas
df = pd.read_excel(excel_file_path,
                   header=0,  # Use the first row as column names
                   index_col=None,  # Do not use any column as index
                   usecols=None  # Read all columns
                   )

# Display the DataFrame
print(df)
```
